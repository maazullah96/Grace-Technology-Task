```python
import requests
from PIL import Image
import matplotlib.pyplot as plt

new_size = (800, 600)  # Size for resized images
images_per_row = 2  # Number of images to display side by side
total_images = 10  # Total number of images to display

# Create an empty list to store resized images
resized_images = []

# Fetch and resize the images
for _ in range(total_images):
    img = requests.get("https://100k-faces.glitch.me/random-image", stream=True)
    orgImg = Image.open(img.raw)
    resized_image = orgImg.resize(new_size)
    resized_images.append(resized_image)
    image = pipe(prompt=prompt, image=resized_image, negative_prompt=None, strength=1).images[0]
    resized_images.append(image)
```

```python

# Create a figure and specify the number of rows and columns for the grid
fig, axes = plt.subplots(nrows=10 , ncols=images_per_row, figsize=(10, 8))

# Iterate over the axes and display images
for ax, image in zip(axes.flat, resized_images):
    ax.imshow(image)
    ax.axis('off')

# Adjust the spacing between subplots
# plt.tight_layout()

# Show the plot
plt.show()
```
